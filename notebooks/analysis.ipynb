# Primetrade Analysis
This notebook contains the analysis workflow for the Bitcoin sentiment and Hyperliquid trader performance assessment.

```python
# 1. Import Libraries
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
# 2. Load Data
fear_greed = pd.read_csv("../data/fear_greed_index.csv")
trades = pd.read_csv("../data/historical_data.csv")
```

```python
# 3. Data Cleaning
fear_greed['Date'] = pd.to_datetime(fear_greed['Date'])
trades['time'] = pd.to_datetime(trades['time'])
```

```python
# 4. Merge Datasets
trades['trade_date'] = trades['time'].dt.date
trades['trade_date'] = pd.to_datetime(trades['trade_date'])
merged = trades.merge(
    fear_greed,
    left_on='trade_date',
    right_on='Date',
    how='left'
)
```

```python
# 5. Feature Engineering
merged['win_trade'] = np.where(
    merged['closedPnL'] > 0,
    1,
    0
)
merged['trade_value'] = (
    merged['execution price'] *
    merged['size']
)
```

```python
# 6. Analysis
avg_pnl = merged.groupby(
    'Classification'
)[ 'closedPnL'].mean()
win_rate = merged.groupby(
    'Classification'
)[ 'win_trade'].mean() * 100
```
